```python
import torch
import timm
```

Let's load the ConvNeXt V2 Tiny model and inspect its parameters.

```python
# Load the ConvNeXt V2 Tiny model
model = timm.create_model('convnextv2_tiny', pretrained=True)

# Calculate the number of parameters
num_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"ConvNeXt V2 Tiny model has {num_params:,} trainable parameters.")
```
